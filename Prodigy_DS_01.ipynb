4.Data Cleaning

```python
#extract top ten countries from atotal population
total_population_data = data[data["Series Code"] == "SP.POP.TOTL"]
total_population_sorted = total_population_data.sort_values(by="2022", ascending=False)
total_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries of total population")
print(total_top_ten_countries )
```

5.Data Visualization

```python
def add_value_labels(ax, spacing=10):
    for rect in ax.patches:
        x = rect.get_x() + rect.get_width() / 2
        y = rect.get_height()-3
        va = 'bottom' if y >= 0 else 'top'
        label = "{}".format(y)
        y_shift = spacing * (1 if y >= 0 else -1)
        ax.annotate(label, (x, y), xytext=(0, y_shift),textcoords="offset points", ha='center', va=va)
```

Top ten countries of total population in year 2020 and 2015

```python
# Create the bar plot
plt.figure(figsize=(20, 16))
plt.rcParams['axes.facecolor'] = 'black'

plt.subplot(2,2,1)
plot=sns.barplot(x="2020", y="Country Name", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2020)")
plt.xlabel("Total Population")
plt.ylabel("Country")
add_value_labels(plot)

plt.subplot(2,2,2)
plot=sns.barplot(x="2015", y="Country Name", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top Ten Countries with Total Population (2015)")
plt.xlabel("Total Population")
plt.ylabel("Country")
add_value_labels(plot)

plt.tight_layout()
```
